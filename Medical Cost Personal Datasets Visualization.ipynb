```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### spliting dataset based on age 

```python
aged_ppl=df[df['age']>=60]
normal_age=df[df['age']<60]
```

#### Univariate categorical analysis

```python
ax=sns.countplot(data=aged_ppl,x='sex',hue='smoker')
plt.title('Aged people insurance taken ratio with gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
ax=sns.countplot(data=normal_age,x='sex',hue='smoker')
plt.title('Normal age people insurance taken ratio with gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
ax=sns.countplot(data=df,x='region')
plt.title("Beneficiary's residential areawise distribution in the US")
for bar in ax.containers:
    ax.bar_label(bar)
```

#### Univariate numerical analysis

```python
sns.histplot(data=df,x='age')
plt.title('Insurance taken ratio based on Age')
```

```python
sns.countplot(data=df,x='children')
plt.title('Number of children covered in insurance')
```

```python
sns.histplot(data=df,x='bmi')
plt.title('Insurance taken ratio based on BMI')
```

#### Bivariate analysis

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df,x='age',y='charges')
plt.title('Individual medical costs billed by insurance based on Age')
```

```python
plt.figure(figsize=(15,7))
sns.barplot(data=df,x='sex',y='age',hue='region')
```
